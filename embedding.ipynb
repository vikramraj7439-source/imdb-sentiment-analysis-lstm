```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent = [
    "The glass of milk",
    "The glass of juice",
    "the glass of milk",
    "the cup of tea",
    "I am a good boy",
    "understand the meaning of words"
]
```

```python
## Define the vocabulary size
vocab_size = 10000
```

```python
### One Hot Representation 
one_hot_representations = [one_hot(words, vocab_size) for words in sent]
one_hot_representations
```

```python
## word Embedding Representation

from tensorflow.keras.layers import Embedding
# from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length=8
embedded_docs=pad_sequences(one_hot_representations,padding='pre',maxlen=sent_length)
print(embedded_docs) # pre means 0's at first
```

```python
## feature representation
dim=10
```

```python
model=Sequential()
model.add(Embedding(vocab_size,dim,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
